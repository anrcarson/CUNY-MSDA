<h2> Solution </h2>
I will be using <i>networkx</i> for my package of choice and will manually recreate the Krackhardt kite shown.

```python
#import packages needed for network creation
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#create graph object
G = nx.Graph() 
```

```python
#create nodes manually
nodes = ['Andre','Beverly','Carol','Diane','Ed','Fernando','Garth','Heather','Ike','Jane']
G.add_nodes_from(nodes)
```

```python
#create edges manually
edges = [('Andre','Beverly'),('Andre','Diane'),('Andre','Carol'), ('Andre','Fernando'),
        ('Beverly','Ed'),('Beverly','Diane'), ('Beverly','Garth'),
        ('Carol','Diane'), ('Carol','Fernando'),
        ('Diane','Garth'),('Diane','Fernando'), ('Diane','Ed'),
         ('Ed','Garth'),
        ('Fernando','Garth'),('Fernando','Heather'),
         ('Garth','Heather'),
         ('Heather','Ike'),
         ('Ike','Jane')
        ]
G.add_edges_from(edges)
```

```python
#draw graph. rerun cell till it looks close to a kite.
nx.draw_spring(G, with_labels = True)
```

<h2> Observations </h2>

Diane has the most connections at 6, so she is a key player in this network.  Heather, while only having 3 connections, is a gateway to Ike, who is a gateway to Jane.  So Heather is also important, as most of the members can be reached in multiple ways, but only Jane can be reached through Ike, and Ike through Heather.

Interestingly, Diane does not connect to Heather, and has to go through Fernando or Garth.  So this network could be thought of as two sub graphs: the one that contains all of Diane's connections, and the one containing Heather, Ike, and Jane.  Garth and Fernando are the keys to going back and forth between these two sub networks.

One could imagine that Jane is the boss, Ike reports to Jane, and Heather reports to Ike.  Both Fernando and Garth report to Heather as managers, and perhaps Diane is the team admin.
